```python
def class_to_string(df):
    
    if df == 1:
        new_df = '1st'
    elif df == 2:
        new_df = '2nd'
    else:
        new_df = '3rd'
    return new_df

# Apply the function to the titanic_data, creating a 'Class' column:

titatincdf['Class'] = titatincdf['Pclass'].apply(class_to_string)

# Compute relative survival rates, by class

total_first = titatincdf.groupby('Class').count()['PassengerId']['1st']
survived_first = titatincdf.groupby('Class').sum()['Survived']['1st']
first_survival = survived_first * 100 / total_first

total_second = titatincdf.groupby('Class').count()['PassengerId']['2nd']
survived_second = titatincdf.groupby('Class').sum()['Survived']['2nd']
second_survival = survived_second * 100 / total_second

total_third = titatincdf.groupby('Class').count()['PassengerId']['3rd']
survived_third = titatincdf.groupby('Class').sum()['Survived']['3rd']
third_survival = survived_third * 100 / total_third

```
